### Dataset Analysis

```python
# Importing Libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Setting Seaborn theme
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10,6)
```

#### RFM Analysis

```python
# Creating RFM DataFrame

rfm_data = {
    'segment': ['Champions', 'Loyal Customers', 'New Customers',
                'At Risk', 'Lost', 'Promising'],
    'customer_count': [21452, 16162, 101, 6156, 28202, 19773]
}

df_rfm = pd.DataFrame(rfm_data)
df_rfm.head()
```

```python
# Sorting Values for Bar Chart
df_rfm = df_rfm.sort_values(by='customer_count', ascending=True)

# Generating Bar Chart
plt.figure(figsize=(10,6))
bars = plt.barh(df_rfm['segment'], df_rfm['customer_count'], color='#2c7fb8')

plt.xlabel('Number of Customers', fontsize=12)
plt.ylabel('Customer Segments', fontsize=12)
plt.title('Number of Customers by Customer Segments', fontsize=14, pad=15)

plt.show()
```

#### Cohort Analysis

```python
# Creating Cohort DataFrame

cohort_data = {
    'cohort_month': ['2024-01', '2024-02', '2024-03'],
    'pct_30d': [9.70, 10.18, 9.42],
    'pct_60d': [8.74, 8.88, 9.67],
    'pct_90d': [9.23, 9.30, 10.20],
    'pct_120d': [9.27, 9.58, 9.06],
    'pct_150d': [9.12, 8.81, 9.60],
    'pct_180d': [9.38, 9.73, 9.73]
}

df_cohort = pd.DataFrame(cohort_data)
df_cohort.set_index('cohort_month', inplace=True)
```

```python
# Creating HeatMap

plt.figure(figsize=(10,6))
sns.heatmap(df_cohort, annot=True, cmap='Blues', fmt='.2f',cbar_kws={'label': 'Retention %'}, linewidths=.5)

plt.title('Cohort Analysis (Jan - Mar 2024)', fontsize=14, pad=15)
plt.xlabel('Days Since Signup (30-Day Buckets)', fontsize=12)
plt.ylabel('Cohort Month', fontsize=12)

plt.show()
```
